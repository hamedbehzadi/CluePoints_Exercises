## Introduction
### This notebook implements data preprocessing for the text data. In the following cell, we import required packages.

```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd # for reading and manipulating the dataframes. 
from bs4 import BeautifulSoup # This is a packages parsing html tags 
import re # A powerful tool for matching patterns in strings The re module allows to perform a variety of operations,
#such as searching, splitting, and substituting parts of strings based on patterns.
from sklearn.preprocessing import LabelEncoder # for encoding text to uinuqe values
from sklearn.feature_selection import mutual_info_classif # for measuring the mutual information between two components
from sklearn.feature_extraction.text import CountVectorizer # to covert text to embeding
import numpy as np
import matplotlib.pyplot as plt

# Reading train and validation sets
df_list =[ pd.read_csv("stack_overflow_questions_train.csv") ,pd.read_csv("stack_overflow_questions_valid.csv")]
```

### Data Preprocessing

#### Following imported the packages we apply a set of preprocessing operation as follows.

 1- Removing html tags, \t and \n and double spaces. As can be seen in the dataset, unrequired characters are htmls tags, tabs, new lines, and double spaces. 

 2- Merging Title and Body coulmns into new column named Text. 

 3- Appling encoding on the label column. The label columns includes three type levels of the text that indicates an ordering among the values, (1) Low Quality with negative score, (2) Low-quality questions that were closed by the community, and (3) High quality questions. Therefore, we apply uniqu encoding

 4- Dropping ID, Title and Body, Tags and CreationDate columns. Each of this columns will be drop based on a reason. The columns ID and CreationDate does not have any semantical relation with the label. The columns Title and Body are merged into a new columns, then we can drop these columns. Finally, we drop Tags columns due to its poot mutual information with the label column.

```python
# This functions cleans the text by removing HTML tags, tabs, new lines, and double spaces.
def text_cleaning(text):
    # Remove HTML tags
    soup = BeautifulSoup(text, "html.parser")
    text = soup.get_text() # returns text without the HTML tags.
    # Removing tab, newline, and double spaces
    text = re.sub(r'[\t\n\s]+', ' ', text)
    # Striping leading and trailing spaces
    text = text.strip()
    return text

# Over two dataframes, i.e., train and validation datasets, we apply function text_cleaning on each of the columns Title and Body
columns = ['Title','Body']
for df in df_list:
    for column in columns:
        df[column] = df[column].apply(text_cleaning)
        
# List of columns should be drop
columns_to_drop = ['Id', 'Title', 'Body', 'CreationDate']
# when adding new columns as result of merging 'Title' and 'Body', the new columns is added in the lasy layer. We would like to move
#label column named as 'Y' to the last column
column_to_move = 'Y'
# we create an object from the following class to apply encoding.
le = LabelEncoder()
for index,df in enumerate(df_list): #This loop goes over two dataframes
    df['Text'] = df['Title'] + " " + df['Body'] # Merging two columns
    df = df[[col for col in df.columns if col != column_to_move] + [column_to_move]] # Reordering the columns
    df.drop(columns=columns_to_drop, axis=1, inplace=True) # Dropping the columns 'Title' and 'Body'
 
    # target column to be converted into uinuqe encoding
    df['Y'] = le.fit_transform(df['Y'])

    df_list[index] = df
# prininting the columns name of each dataframe to check the correctness of the previous operations
for df in df_list:
    print(df.keys())
```

### Mutual Information

To check whether the column named 'Tags' has any contribution on the label column, we compute the measuring mutual information between Tags column and the label column. 

```python
# Preprocess Tags using CountVectorizer
vectorizer = CountVectorizer(tokenizer=lambda x: x.split(), binary=True)
X = vectorizer.fit_transform(df_list[0]['Tags'])
mi = mutual_info_classif(X, df_list[0]['Y'], discrete_features=True) # Computing Mutual Information
```

```python
#Plotting the MI scores. X-axis shows each of the Tags, and Y-axis shows the MI scores between a given Tag and the label column
x = np.arange(len(mi))
plt.figure()
plt.plot(x,mi,'r-.')
plt.xlabel('Tags')
plt.ylabel('MI')
plt.ylim([0,0.01])
plt.title('MI between each tag and output label')
plt.show()
```

### Removing Tags column due to its week contribution

```python
for index,df in enumerate(df_list):
    df.drop(columns=['Tags'], axis=1, inplace=True)
    df_list[index] = df

for df in df_list:
    print(df.keys())
```

### Check for data imblacne issue

As the last preprocessing step, we count number of samples per class, to see if there is any imbalanc issue in the data.

```python
class_counts = df_list[0]['Y'].value_counts()
x = np.arange(len(class_counts))
plt.bar(x,class_counts)
plt.xlabel('Classes')
plt.ylabel('Number of Samples')
print(class_counts)
```
